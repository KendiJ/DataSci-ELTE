### PCA & The Curse of Dimensionality
- Use Principal Component Analysis (PCA) to reduce a high-dimensional dataset and visualize it in 2D.
The Data: The Digits Dataset (from sklearn.datasets import load_digits). This dataset has 64 features (an $8 \times 8$ grid of pixels representing handwritten numbers).
- Core Tasks:Load the dataset and standardize it.Apply PCA to reduce the 64 features down to just 2 Principal Components.Print the explained_variance_ratio_ to see exactly how much information you preserved.Create a scatter plot using those 2 new components as your X and Y axes, coloring the dots by their actual digit label. You will visually see the clusters form without having to train a predictive model!

1. Imports
2. Load digits
3. Standardize
4. PCA to 2D
5. Variance explained
6. Scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
digits = load_digits()

X = digits.data
y = digits.target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
print("Explained variance ratio:")
print(pca.explained_variance_ratio_)

print("Total preserved variance:", np.sum(pca.explained_variance_ratio_))
```

```python
plt.figure(figsize=(10,7))
scatter = plt.scatter(X_pca[:,0], X_pca[:,1], c=y)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA of Digits Dataset")
plt.colorbar(scatter)
plt.show()
```
